```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def show_slices(slices, titles):
    """Function to display row of image slices"""
    fig, axes = plt.subplots(1, len(slices))
    for i, (slice, title) in enumerate(zip(slices, titles)):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
        axes[i].set_title(title)
        axes[i].set_axis_off()
    plt.show()

# Define the paths to the NIfTI files for each modality
subject_path = '/home/ioj24/Desktop/Multimodal Image Segmentation/Datasets/BraTS2020_TrainingData/MICCAI_BraTS2020_TrainingData/BraTS20_Training_001'
t1_path = os.path.join(subject_path, 'BraTS20_Training_001_t1.nii')
t1gd_path = os.path.join(subject_path, 'BraTS20_Training_001_t1ce.nii')
t2_path = os.path.join(subject_path, 'BraTS20_Training_001_t2.nii')
t2_flair_path = os.path.join(subject_path, 'BraTS20_Training_001_flair.nii')

# Load the image data using nibabel
t1_img = nib.load(t1_path)
t1gd_img = nib.load(t1gd_path)
t2_img = nib.load(t2_path)
t2_flair_img = nib.load(t2_flair_path)

# Get the data arrays from the image objects
t1_data = t1_img.get_fdata()
t1gd_data = t1gd_img.get_fdata()
t2_data = t2_img.get_fdata()
t2_flair_data = t2_flair_img.get_fdata()

# Define the slice index you want to visualize
slice_idx = 50

# Get the slices for each modality
t1_slice = t1_data[:, :, slice_idx]
t1gd_slice = t1gd_data[:, :, slice_idx]
t2_slice = t2_data[:, :, slice_idx]
t2_flair_slice = t2_flair_data[:, :, slice_idx]

# Show the slices using the function defined earlier
show_slices([t1_slice, t1gd_slice, t2_slice, t2_flair_slice],
            ["T1", "T1Gd", "T2", "T2-FLAIR"])
```
